## Stack을 통해 사용하는 기능

1.  `push`: 데이터를 스택에 추가한다.(스택의 하단부터 상단으로 차곡차곡
    쌓는다.
2.  `pop`: 스택의 최상단 데이터를 삭제한다.
3.  `top`: 스택에 데이터가 몇 개 들어있는지 확인한다.

여러 개의 쇠막대기를 레이저로 절단하려고 한다. 효율적인 작업을 위해서

쇠막대기들을 자른다. 쇠막대기와 레이저의 배치는 다음 조건을 만족한다.

-   쇠막대기는 자신보다 긴 쇠막대기 위에만 놓일 수 있다.

        끝점은 겹치지 않도록 놓는다.
-   각 쇠막대기를 자르는 레이저는 적어도 하나 존재한다.
-   레이저는 어떤 쇠막대기릐 양 끝점과도 겹치지 않는다.

## 크게 만들기

N 자리 숫자가 주어졌을때, 여기서 숫자 K개를 얻을 수 있는 가장 큰 수를
구하는 프로그램을 작성하시오.

`-` 맨 왼쪽에 있는 수가 가장 큰 수를 선택했을때 결과도 크겠지?

`-` array보다는 stack으로 시간복잡도를 줄일 수 있다.

```python
n, k = map(int, input().split())
number = list(input())

answer = []
cnt = k
for num in number: 
    while answer and cnt>0 and answer[-1] <num: 
        # 만약 answer에 값이 있고 지울 수 있는 수 k가 남아있고, answer의 마지막 값이 num보다 작다면?
        del answer[-1]
        # answer 마지막 값 삭제하고
        cnt-=1
        # 지울 수 있는 k 수도 줄인다.
    answer.append(num)
    
print(''.join(answer[:n-k]))
```
