## (1)

가설 $H_0 : \sigma^4 = 4$ vs $H_1 : \sigma^2 = 9$에 대한 최강력 검정의
기각역은

$$C = \{ (x_1, \dots, x_n) : \sum^n_{i=1} x^2_i \ge c \}$$

의 꼴로 주어짐을 보이시오.

**answer**

$L(\sigma^2) = \Pi^n_{i=1} f(x_2 : \sigma^2) = \Pi^n_{i=1} \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{x_i^2}{2\sigma^2}} * \mu=0$

$= (\frac{1}{2\pi\sigma^2})^{\frac{n}{2}} e^{-\frac{\sum^n_{i=1}x_i^2}{2\sigma^2}}$

네이만 피어슨 정의에 의하면,
$LR = \frac{L(H_0)}{L(H_1)} = \frac{L(4)}{L(9)} \le k$

$LR = \frac{L(4)}{L(9)} = \frac{(\frac{1}{2\pi 4})^{\frac{n}{2}} e^{-\frac{\sum^n_{i=1} x_i^2}{2 \times 4}}}{(\frac{1}{2\pi 9})^{\frac{n}{2}} e^{-\frac{\sum^n_{i=1} x_i^2}{2 \times 9}}}$

$= (\frac{9}{4})^{\frac{n}{2}}e^{-\sum^n_{i=1}x^2_i(\frac{1}{8}-\frac{1}{18})}$

$= (\frac{9}{4})^{\frac{n}{2}}e^{-\frac{5}{72}\sum^n_{i=1}x^2_i} \le k$

$\to e^{-\frac{5}{72}\sum^n_{i=1}x^2_i} \le k$

$\to -\frac{5}{72}\sum^n_{i=1}x^2_i \le k$

$\to \sum^n_{i=1}x^2_i \ge k$

기각역:
$\therefore c = \{ (x_1, \dots ,x_n) : \sum^n_{i=1} x^2_i \ge c)$

$= P(\sum^{20}_{i=1} \frac{\chi^2_i}{4} \ge \frac{k}{4} | \sigma^2 = 4)$

$c = \{ x_{(n)} \le c\text{ 상수}\}$

$c = \{ X_{(n)} \le c \}$

$\alpha = P(X_{(n)} \le c | \theta = \theta_0 \} = \int^c_{-\infty} f_n (x : \theta_0)dx$

# 23.

확률밀도함수가 $f(x;0)=\theta x^{\theta-1}I(0<x<1)$인 모집단으로부터
랜덤표본 $X_1,X_2,\dots,X_n$을 추출하여 $H_0 : \theta = \theta_0$ vs
$H_1:\theta \neq \theta_0$의 가설검정을 하려고 한다. 이 때 균일 최강력
검정이 존재하지 않음을 보여라.

**answer**

-   $H_0 : \theta = \theta_0$ vs
    $H_1 : \theta = \theta_1 (\theta_1 < \theta_0)$

네이만 피어슨 정리
$c = \{ (x_1,\dots,x_n):\frac{L(\theta_0)}{L(\theta_1)}\le k \}$

$L(\theta) = \Pi^n_{i=1}f(x:\theta) = \Pi^n_{i=1}\theta x^{\theta-1}_i I(0<x_i < 1)$

$\frac{L(\theta_0)}{L(\theta_1)} = \frac{\Pi^n_{i=1} \theta_0 x_i^{\theta_0 - 1} I(0<x_i<1)}{\Pi^n_{i=1} \theta_1 x_i^{\theta_1-1}I(0<x_i<1)} = (\frac{\theta_0}{\theta_1})^n (\Pi^n_{i=1}x_i)^{\theta_0-\theta_1} \le k$

$\to (\Pi^n_{i=1}x_i)^{\theta_0 - \theta_1} \le k$

$\to (\theta_0 - \theta_1) log\Pi^n_{i=1} x_i \le k$

$\to (\theta_0 - \theta_1)\Pi^n_{i=1} x_i \le k$

$\to \sum^n_{i=1} x_i \le k$ 만족하는 표본의 영역이 기각역이다. $\sim$
최강력 기각역

$\to \theta_1$과 무관하게 결정되는 기각역

$\therefore$ 이 기각역은
$H_0:\theta = \theta_0, H_1 : \theta_1 < \theta_0$의 균일 최강력
기각역이다.

-   $H_0 : \theta = \theta_0$ vs
    $H_1 : \theta = \theta_1 (\theta_1 > \theta_0)$

$\frac{L(\theta_0)}{L(\theta_1)} = \frac{\Pi^n_{i=1} \theta_0 x_i^{\theta_0 - 1} I(0<x_i<1)}{\Pi^n_{i=1} \theta_1 x_i^{\theta_1-1}I(0<x_i<1)} = (\frac{\theta_0}{\theta_1})^n (\Pi^n_{i=1}x_i)^{\theta_0-\theta_1} \le k$

$\to (\Pi^n_{i=1}x_i)^{\theta_0 - \theta_1} \le k$

$\star \theta_0<\theta_1$이라 부등호 바뀜

$\to (\theta_0 - \theta_1)\Pi^n_{i=1} x_i \ge k$

$\to \sum^n_{i=1} x_i \ge k$ 만족하는 표본의 영역이 기각역이다.

$\to \theta_1$과 무관하게 결정되는 기각역

# 32.

$X \sim B(100,p_x),Y \sim B(106,p_y)$일 때 가설 $H_0 : p_x = p_y$ vs
$H_1 : p_x \neq p_y$ 를 고려해보자. 실험을 통하여 $X=40, Y=68$을
얻었다고 할 때 정규근사를 활용하여 유의수준 $\alpha = 0.05$인 일반화
가능도비 검정법으로 검정하시오.

**answer**

$L(p_x,p_y) = f(x,y:p_x,p_y)$ 서로 독립이다.문제에 언급은 없다.

$= f_x(x:p_x)f_y(y:p_y)= \begin{pmatrix} 100 \\ x\end{pmatrix} p_x^x(1-p_x)^{100-x} \begin{pmatrix} 106 \\ y\end{pmatrix} p_y^y(1-p_y)^{106-x}$

-   $\Omega_0 = \{(p_x,p_y): 0<p_x=p_y<1\}$
-   $\Omega = \{(p_x,p_y): 0<p_x<1,0<p_y<1\}$

$\frac{\partial log L(p_x,p_y)}{\partial p_x} = 0 \to x log p_x + (100-x) log (1-p_x) + c \to \hat{p_x} \to \frac{x}{p_x} - \frac{(100-x)}{(1-p_x)} + c =0\to \hat{p_x} = \frac{x}{100}$

$\frac{\partial log L (p_x, p_y)}{\partial p_y} = 0 \to \hat{p}_y = \frac{y}{106}$

$p_x = p_y = p$,
$L(p) = \begin{pmatrix} 100 \\ x\end{pmatrix} p_x^x(1-p_x)^{100-x} \begin{pmatrix} 106 \\ y\end{pmatrix} p_y^y(1-p_y)^{106-x} = \begin{pmatrix} 100 \\ x\end{pmatrix} p^{x+y}(1-p)^{206-x-y} \begin{pmatrix} 106 \\ y\end{pmatrix}$

$\frac{\partial log L(p)}{\partial p} = 0 = \frac{\partial(x+y) log p + (106 -x-y)log(1-p) + c}{\partial p} \to \hat{p}_0 = \frac{x+y}{206}$

$\Lambda = \frac{L(\hat{p}_0)}{L(\hat{p}_x,\hat{p}_y)} = \frac{\begin{pmatrix} 100 \\ x\end{pmatrix}\begin{pmatrix} 106 \\ y\end{pmatrix}(\frac{x+y}{206})^{x+y}(1-\frac{x+y}{206})^{206-x-y}}{\begin{pmatrix} 100 \\ x\end{pmatrix} (\frac{x}{100})^x (1-\frac{x}{100})^x)^{100-x}\begin{pmatrix} 106 \\ y\end{pmatrix}(\frac{y}{106})^y(1-\frac{y}{106})^{106-y}} \le k$

$\to |\frac{\frac{x}{100} - \frac{y}{106}}{\sqrt{206 - \hat{p}_0(1-\hat{p}_0)/10600}}| \ge Z_{0.025}$

$\hat{p}_0 = \frac{x+y}{206}$, 문제에서 제시된 값 대입 $X=40,Y=68$

$\to \frac{\frac{40}{100} - \frac{68}{106}}{\sqrt{206 - \frac{108}{206}(1- \frac{108}{206})/10600}} = 3.47$

$|3.47| > 1.96 (Z_{0.025}) \therefore$ 귀무가설을 기각할 수 있다.
